```python
from faker import Faker
import random
from datetime import datetime

# Initialize Faker and seed random number generator for reproducibility
fake = Faker()
random.seed(42)

# Function to generate non-redundant fake dates between 1990 and 2010
def generate_non_redundant_dates(num_dates):
    dates = set()
    while len(dates) < num_dates:
        fake_date = fake.date_time_between_dates(datetime(1990, 1, 1), datetime(2010, 12, 31))
        dates.add(fake_date.strftime('%Y-%m-%d'))
    return list(dates)

# Generate 100 non-redundant fake dates between 1990 and 2010
fake_dates = generate_non_redundant_dates(1000)

#Print the generated fake dates
for date in fake_dates:
    print(date)
len(set(fake_dates))
```

```python
import pandas as pd
from faker import Faker
import random


fake = Faker()

# Define the number of sales records you want
num_sales = 800

# Generate sales records
sales_data = []

for _ in range(num_sales):
    sale_date = fake.date_between(start_date='-1y', end_date='today')
    sales_price = round(random.uniform(50, 150), 2)  # Random sales price between $5 and $20
    sales_data.append((sale_date,  sales_price))

# Create a DataFrame
sales_df = pd.DataFrame(sales_data, columns=['Sale Date', 'Sales Price'])

# Display the DataFrame
sales_df.head()
sales_df.shape
```
